# Analisando a Violência no Rio de Janeiro

Apesar de ser apenas o 13º estado mais violento do Brasil o Rio de Janeiro chama atenção por ser o segundo estado mais violento da região central do país, além disso o Rio está entre os 17 estados com média de violência maior que a própria media do País. (Fonte: Gazeta do Povo)
<br><br>
Se levarmos em consideração a capital do estado que é a segunda maior cidade do país com 6,2 milhões de habitantes, a metrópole que é principal destino de turistas estrangeiros é semanalmente destaque em noticiários por causa de violência. A violência urbana no município esta associada a tráfico de drogas, guerras entre facções por comando territorial, assaltos, entre outros.

## Informações Estatísticas da Violência no Rio de Janeiro

```
#Resumo estatístico do Dataframe
df.describe()
```

```
#Média da coluna roubo_veiculo
mdrv = df.roubo_veiculo.mean()

#Média da coluna furto_veiculos
mdfv = df.furto_veiculos.mean()

#Média da coluna furto_veiculos
mdrve = df.recuperacao_veiculos.mean()

print('Média da coluna roubo_veiculo: ', mdrv)
print('Média da coluna furto_veiculos: ', mdfv)
print('Média da coluna reculperação_veiculos: ', mdrve)
```

```
# Calculo da porcentagem de carros reculperados em relação a carros roubados 
#+ carros furtados
df.recuperacao_veiculos.sum() / (df.roubo_veiculo.sum() + df.furto_veiculos.sum())
```

## **Observações:**

Através dos dados acima é possível notar que em média nem metade dos veículos roubados e furtados são recuperados, apenas 42,98% dos veículos foram recuperados. 

## Visualização de Dados

Posteriormente iremos representar algumas informações graficamente, essa etapa é muito importante, pois, gráficos são ferramentas que facilitam a análise e interpretação de um conjunto de dados. ... Sua importância está ligada à facilidade e rapidez com que podemos interpretar as informações. Os dados coletados podem ser organizados em gráficos e apresentados de uma forma mais clara e objetiva.

A seguir iremos *plotar* as seguintes informações:

* Histograma de `hom_doloso`
* Gráfico de linhas para a variável `roubo_em_coletivo`


```
# Histograma da coluna hom_doloso
df.hom_doloso.hist();
```

```
# Média e Mediana da coluna hom_doloso
print("Média hom_doloso: {0}".format(df.hom_doloso.mean()))
```

## **Observações:**
O histograma tem como objetivo mostrar a distribuição de frequências de dados obtidos por medições, para identificarmos a frequência com que algo acontece, nesse caso foi possível comprovar, junto com a média, que a mior frequência de dados está perto de 500.

```
#Grafico de linhas para coluna roubo_em_coletivo
df['roubo_em_coletivo'].plot();
```

```
#Grafico de linhas para coluna roubo_em_coletivo relacionando os roubos com o ano
x = df.vano
y = df.roubo_em_coletivo
plt.plot(x, y);
```

```
print("Média de roubo_em_coletivo: {0}".format(df.roubo_em_coletivo.mean()))
```

## **Observações:**
Devido ao gráfico de linhas ser ideal para exibir tendências ao longo do tempo, podemos perceber que após o ano de 2015 o número de roubos em coletivos aumentou muito chegando a quase triplicar a média em alguns meses.
